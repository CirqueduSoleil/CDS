```python
import datetime
from datetime import date

import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
# Importer le fichier Montlhy term provenant de SF
df_mt = pd.read_excel('mt.xls')

#import census dsf
df_dsf = pd.read_excel('dsf.xls')
df_dsf['Employee Number']=df_dsf["NO_PAYE_MEMBRE"]

#import census psw
df_psw = pd.read_excel('PSW.xls')
df_psw['Employee Number']=df_psw['Emp ID #']
df_psw.rename(columns={ df_psw.columns[3]: "Division" }, inplace = True)
#df_psw['Div Name']
#Enlever ceux qui n'ont pas de régime
df_mt = df_mt[pd.notnull(df_mt['Saving Plan Type'])]

#Créer un df pour régime canadien et un US

df_mt_can1=df_mt.drop(df_mt[(df_mt['Saving Plan Type'] != 'Canadian Pension Plan')&(df_mt['Saving Plan Type'] != 'Executive – Canadian Pension Plan')].index)
df_mt_us1=df_mt.drop(df_mt[(df_mt['Saving Plan Type'] != 'US 401K')&(df_mt['Saving Plan Type'] != 'Executive – US Non-Qualified Plan')].index)

#Garder seulement la ligne avec bon SSN
df_mt_can1=df_mt_can1.drop(df_mt_can1[(df_mt_can1['National ID - Country'] != 'CAN')].index)
df_mt_us1=df_mt_us1.drop(df_mt_us1[(df_mt_us1['National ID - Country'] != 'USA')].index)

################################################### CANADIEN #############################################################

#Ajouter sous groupe
df_mt_can1 = pd.merge(df_mt_can1,df_dsf[['Employee Number',"SOUS_GROUPE"]],on='Employee Number', how='left')


df_mt_can1=df_mt_can1.drop_duplicates(subset ="Employee Number" )

#Enlever les décès (ne sont pas traités de la même façon)
df_mt_can1=df_mt_can1.drop(df_mt_can1[(df_mt_can1['Employee Status'] == 'Deceased')].index)

#Mettre les termes de DSF
mask = df_mt_can1['Employee Status'].eq('Terminated')
df_mt_can1.loc[mask , 'Employee Status'] = "Cessation d'emploi"
mask = df_mt_can1['Employee Status'].eq('Retired')
df_mt_can1.loc[mask , 'Employee Status'] = "Retraite"


df_mt_can=pd.DataFrame()
#Template Canadien
df_mt_can['Employee Number']=df_mt_can1['Employee Number']
df_mt_can.insert(0, "Transaction", 'D', True)
df_mt_can.insert(1, "Date D'effet", '', True)
df_mt_can.insert(2, "No du groupe", 'G004922', True)
df_mt_can['Sous groupe']=df_mt_can1['SOUS_GROUPE']
df_mt_can.insert(4, "Date D'effet sous groupe", '', True)

df_mt_can.insert(6, "No participant", '', True)
df_mt_can['Last Name']=df_mt_can1['Last Name']
df_mt_can['First Name']=df_mt_can1['First Name']
df_mt_can.insert(9, "Date de naissance", '', True)
df_mt_can['NAS']=df_mt_can1['National ID']
df_mt_can.insert(11, 'Sexe', '', True)
df_mt_can.insert(12, 'Langue', '', True)
df_mt_can.insert(13, 'Statut civil', '', True)
df_mt_can.insert(14, 'Date entree en service', '', True)
df_mt_can['Date de cessation']=df_mt_can1['Last Date Worked']
df_mt_can['Raison cessation']=df_mt_can1['Employee Status']
df_mt_can['Date derniere cotisation']=df_mt_can1['Last Date Worked']+ datetime.timedelta(days=30)
df_mt_can['Province emploi']=''
df_mt_can['Province de residence']=''
df_mt_can['Adresse ligne 1']=''
df_mt_can['Adresse ligne 2']=''
df_mt_can['Adresse ligne 3']=''
df_mt_can['Ville']=''
df_mt_can['Pays']=''
df_mt_can['Code postal']=''
df_mt_can['Telephone domicile']=''
df_mt_can['Telephone travail']=''
df_mt_can['Poste travail']=''
df_mt_can['Mobile']=''
df_mt_can['Courriel']=''
df_mt_can['Nom du produit_1']=''
df_mt_can['Date de participation au produit_1']=''
df_mt_can['Nom du produit_2']=''
df_mt_can['Date de participation au produit_2']=''
df_mt_can['Last work day Year']=df_mt_can['Date de cessation'].dt.year
df_mt_can['Last work day Month']=df_mt_can['Date de cessation'].dt.month
df_mt_can=df_mt_can.drop(df_mt_can[(df_mt_can['Last work day Year'] != 2020)].index)

# par mois de terminaison
df_mt_can_avril=df_mt_can.drop(df_mt_can[(df_mt_can['Last work day Month'] != 4)].index)
df_mt_can_avril=df_mt_can_avril.drop(columns=['Last work day Year', 'Last work day Month'])

df_mt_can_mai=df_mt_can.drop(df_mt_can[(df_mt_can['Last work day Month'] != 5)].index)
df_mt_can_mai=df_mt_can_mai.drop(columns=['Last work day Year', 'Last work day Month'])

df_mt_can_juin=df_mt_can.drop(df_mt_can[(df_mt_can['Last work day Month'] != 6)].index)
df_mt_can_juin=df_mt_can_juin.drop(columns=['Last work day Year', 'Last work day Month'])

df_mt_can_juillet=df_mt_can.drop(df_mt_can[(df_mt_can['Last work day Month'] != 7)].index)
df_mt_can_juillet=df_mt_can_juillet.drop(columns=['Last work day Year', 'Last work day Month'])

df_mt_can_aout=df_mt_can.drop(df_mt_can[(df_mt_can['Last work day Month'] != 8)].index)
df_mt_can_aout=df_mt_can_aout.drop(columns=['Last work day Year', 'Last work day Month'])

df_mt_can_sept=df_mt_can.drop(df_mt_can[(df_mt_can['Last work day Month'] != 9)].index)
df_mt_can_sept=df_mt_can_sept.drop(columns=['Last work day Year', 'Last work day Month'])

df_mt_can_oct=df_mt_can.drop(df_mt_can[(df_mt_can['Last work day Month'] != 10)].index)
df_mt_can_oct=df_mt_can_oct.drop(columns=['Last work day Year', 'Last work day Month'])

df_mt_can_nov=df_mt_can.drop(df_mt_can[(df_mt_can['Last work day Month'] != 11)].index)
df_mt_can_nov=df_mt_can_nov.drop(columns=['Last work day Year', 'Last work day Month'])

df_mt_can_dec=df_mt_can.drop(df_mt_can[(df_mt_can['Last work day Month'] != 12)].index)
df_mt_can_dec=df_mt_can_dec.drop(columns=['Last work day Year', 'Last work day Month'])

#display(df_mt_can_avril)
####################################################### US ####################################################################
#Enlever Doublons
df_mt_us1=df_mt_us1.drop_duplicates(subset ="Employee Number" )

#Mettre les termes de PSW
mask = df_mt_us1['Employee Status'].eq('Terminated')
df_mt_us1.loc[mask , 'Employee Status'] = "T"
mask = df_mt_us1['Employee Status'].eq('Retired')
df_mt_us1.loc[mask , 'Employee Status'] = "R"
df_mt_us1['Partic SSN']=df_mt_us1['National ID']

#Ajouter sous groupe
df_mt_us1 = pd.merge(df_mt_us1,df_psw[['Employee Number','Division']],on='Employee Number', how='left')

# Template US
df_mt_us=pd.DataFrame()
df_mt_us['Division Code']=df_mt_us1['Division']
df_mt_us['SSN']=df_mt_us1['Partic SSN']
df_mt_us['Term Date']=df_mt_us1['Last Date Worked']
df_mt_us['Status Code']=df_mt_us1['Employee Status']



df_mt_us['Last work day Year']=df_mt_us['Term Date'].dt.year
df_mt_us['Last work day Month']=df_mt_us['Term Date'].dt.month
df_mt_us=df_mt_us.drop(df_mt_us[(df_mt_us['Last work day Year'] != 2020)].index)

# par mois de terminaison
df_mt_us_avril=df_mt_us.drop(df_mt_us[(df_mt_us['Last work day Month'] != 4)].index)
df_mt_us_avril=df_mt_us_avril.drop(columns=['Last work day Year', 'Last work day Month'])

df_mt_us_mai=df_mt_us.drop(df_mt_us[(df_mt_us['Last work day Month'] != 5)].index)
df_mt_us_mai=df_mt_us_mai.drop(columns=['Last work day Year', 'Last work day Month'])

df_mt_us_juin=df_mt_us.drop(df_mt_us[(df_mt_us['Last work day Month'] != 6)].index)
df_mt_us_juin=df_mt_us_juin.drop(columns=['Last work day Year', 'Last work day Month'])

df_mt_us_juillet=df_mt_us.drop(df_mt_us[(df_mt_us['Last work day Month'] != 7)].index)
df_mt_us_juillet=df_mt_us_juillet.drop(columns=['Last work day Year', 'Last work day Month'])

df_mt_us_aout=df_mt_us.drop(df_mt_us[(df_mt_us['Last work day Month'] != 8)].index)
df_mt_us_aout=df_mt_us_aout.drop(columns=['Last work day Year', 'Last work day Month'])

df_mt_us_sept=df_mt_us.drop(df_mt_us[(df_mt_us['Last work day Month'] != 9)].index)
df_mt_us_sept=df_mt_us_sept.drop(columns=['Last work day Year', 'Last work day Month'])

df_mt_us_oct=df_mt_us.drop(df_mt_us[(df_mt_us['Last work day Month'] != 10)].index)
df_mt_us_oct=df_mt_us_oct.drop(columns=['Last work day Year', 'Last work day Month'])

df_mt_us_nov=df_mt_us.drop(df_mt_us[(df_mt_us['Last work day Month'] != 11)].index)
df_mt_us_nov=df_mt_us_nov.drop(columns=['Last work day Year', 'Last work day Month'])

df_mt_us_dec=df_mt_us.drop(df_mt_us[(df_mt_us['Last work day Month'] != 12)].index)
df_mt_us_dec=df_mt_us_dec.drop(columns=['Last work day Year', 'Last work day Month'])

display(df_mt_us)
```
